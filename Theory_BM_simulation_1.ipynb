```python
from aleatory.processes import BrownianMotion
import matplotlib.pyplot as plt
import numpy as np
```

# 1-D Brownian Motion

```python
brownian = BrownianMotion()
brownian.draw(n=100, N=1,figsize=(8,5),marginal=False)
```

```python
brownian = BrownianMotion(T=1000)
bm_sim = brownian.draw(n=100, N=10000,  figsize=(12,6),envelope=True)
```

# Correlated Brownian Motion

```python
n = 1000; dt = 0.01; rho = 0.8
dW1 = np.random.normal(0, np.sqrt(dt), n)
dW2 = rho * dW1 + np.sqrt(1 - rho **2) * np.random.normal(0, np.sqrt(dt), n)
W1 = np.cumsum(dW1)
W2 = np.cumsum(dW2)


fig = plt.figure(figsize=(8, 5))
#plt.style.use('default')
plt.plot(W1,label ='$X_t ^{(1)}$',color='darkblue')
plt.plot(W2,label = '$X_t ^{(2)}$',color='darkgoldenrod')
plt.axhline(y=0, label ='$E[X_t ^{(1)}] = E[X_t ^{(2)}] = 0$',color='orange', linestyle='dashed',linewidth=2.0)
plt.xlabel('t')
plt.ylabel('X(t)')
plt.title(r'Correlated Brownian motion simulation $X_t ^{(1)},X_t ^{(2)}, t \in [t_0,T]$ with $\rho =0.8$')
plt.legend()
plt.grid()
plt.show()
```
