```python
from aleatory.processes import BrownianMotion
import matplotlib.pyplot as plt
import numpy as np
from fbm import FBM
```

# 1-D Brownian Motion

```python
n = 1000; dt = 0.01
dW1 = np.random.normal(0, np.sqrt(dt), n)

W1 = np.cumsum(dW1)
#W2 = np.cumsum(dW2)


fig = plt.figure(figsize=(10, 4))
#plt.style.use('default')
plt.plot(W1,label ='$W_t ^{(1)}$',color='darkblue')

plt.axhline(y=0, label ='$E[W_t] = 0$',color='orange', linestyle='dashed',linewidth=2.0)
plt.xlabel('t',fontsize=20)
plt.ylabel('$W_t$',fontsize=20)
#plt.title(r'Correlated Brownian motion simulation $X_t ^{(1)},X_t ^{(2)}, t \in [t_0,T]$ with $\rho =0.8$')
plt.legend()
plt.grid()
plt.show()
```

# Correlated Brownian Motion

```python
n = 1000; dt = 0.01; rho = 0.8
dW1 = np.random.normal(0, np.sqrt(dt), n)
dW2 = rho * dW1 + np.sqrt(1 - rho **2) * np.random.normal(0, np.sqrt(dt), n)
W1 = np.cumsum(dW1)
W2 = np.cumsum(dW2)


fig = plt.figure(figsize=(10, 4))
#plt.style.use('default')
plt.plot(W1,label ='$W_t ^{(1)}$',color='darkblue')
plt.plot(W2,label = '$W_t ^{(2)}$',color='darkgoldenrod')
plt.axhline(y=0, label ='$E[W_t ^{(1)}] = E[W_t ^{(2)}] = 0$',color='orange', linestyle='dashed',linewidth=2.0)
plt.xlabel('t',fontsize=20)
plt.ylabel('$W_t$',fontsize=20)
#plt.title(r'Correlated Brownian motion simulation $X_t ^{(1)},X_t ^{(2)}, t \in [t_0,T]$ with $\rho =0.8$')
plt.legend()
plt.grid()
plt.show()
```

```python
f_05 = FBM(n=9999, hurst=0.05, length=1, method='daviesharte')
f_10 = FBM(n=9999, hurst=0.10, length=1, method='daviesharte')
f_30 = FBM(n=9999, hurst=0.3, length=1, method='daviesharte')
f_50 = FBM(n=9999, hurst=0.5, length=1, method='daviesharte')
f_70 = FBM(n=9999, hurst=0.7, length=1, method='daviesharte')

fig = plt.figure(figsize=(10, 4))
#plt.style.use('default')
plt.plot(f_05.fbm(),label ='$H=0.05$',color='blue')
plt.plot(f_10.fbm(),label ='$H=0.10$',color='brown')
plt.plot(f_30.fbm(),label ='$H=0.30$',color='gold')
plt.plot(f_50.fbm(),label ='$H=0.50$',color='purple')
plt.plot(f_70.fbm(),label ='$H=0.70$',color='green')
#plt.axhline(y=0, label ='$E[W_t ^{(1)}] = E[W_t ^{(2)}] = 0$',color='orange', linestyle='dashed',linewidth=2.0)
plt.xlabel('t',fontsize=20)
plt.ylabel('$B_t ^H$',fontsize=20)

plt.legend()
plt.grid()
plt.show()
```
